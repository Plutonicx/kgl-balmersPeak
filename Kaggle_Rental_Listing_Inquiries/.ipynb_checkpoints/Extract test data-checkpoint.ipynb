# Create helper function and manually join these together (ugh!)

## This could be programatically done but this is easier and quicker

```python
def generate_relational_data(item_name):
    items = test_set.get(item_name)
    item_list = [ [i, items[str(i)]] for i in items ]
    item_frame = pd.DataFrame(item_list)
    item_frame.columns = ['ID',item_name]
    return(item_frame)

def generate_relational_data_string(item_name):
    items = test_set.get(item_name)
    item_list = [ [i, items[str(i)]] for i in items ]
    item_frame = pd.DataFrame(item_list)
    item_frame.columns = ['ID',item_name]
    return(item_frame)
```

# Check if features are clean

```python
features = training_set.get('features')
feature_items = [ [i, features[str(i)]] for i in tqdm(features) ]
```

## Get all unique feature items. Storing in a set allows for $ O (1) $ lookup

```python
seen = set()

item_list = []

for feature in tqdm(feature_items):
    for item in feature[1]:
        if item not in seen and not seen.add(item):
            item_list.append(item)
```

```python
test = pd.DataFrame(item_list)
```

## For now: Just find most common items

```python
item_list = []

for feature in tqdm(feature_items):
    for item in feature[1]:
        item_list.append(item)
        
item_frame = pd.DataFrame(item_list)
```

```python
item_frame.columns = ['features']
```

```python
item_counts = pd.DataFrame(item_frame.groupby('features').size().rename('counts'))
```

```python
item_counts.sort_values(by = 'counts', ascending=False)
```

## Definitely need to cluster or atleast regex. Look above: Floor-to-ceiling windows & Floor To Ceiling Windows are seperate items.

## For now just use features that have a count more than 200 (this contains dupes but this is for a first look at modelling the data

```python
item_lists = list(item_counts[item_counts['counts'] >= 200].index)
```

```python
item_lists[4]
```

```python
import re
```

```python
def item_exists_in_feature_list(item, feature_list):
    if re.search(pattern=item,string=feature_list):
        return(1)
    else:
        return(0)
```

```python
print(item_exists_in_feature_list('Dishwasher','[Fireplace, Laundry in Unit, Dishwasher, Hardwood]'))
print(item_exists_in_feature_list('Dishwaser','[Fireplace, Laundry in Unit, Dishwasher, Hardwood]'))
```

```python
feature_matrix_build = []

for feature_list in tqdm(full_test['features']):
    feature_column_build = [item_exists_in_feature_list(item.lower(), str(feature_list).lower()) for item in item_lists]
    feature_matrix_build.append(feature_column_build)
    

```

```python
feature_frame = pd.DataFrame(feature_matrix_build)
feature_frame.columns = item_lists
```

```python
first_df = pd.concat([ full_test, feature_frame], axis = 1)
first_df.to_csv('../../Data/RentalListingData/first_dataframe_test.csv', sep='~')
```
